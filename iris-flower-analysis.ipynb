```python
# Looding The Library with iris dataset
from sklearn.datasets import load_iris
#loading scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier
# Loading Pandas
import pandas as pd
#loading numpy
import numpy as np
# Setting random seed
np.random.seed(0)
```

```python
# print(iris)
```

```python
# Creating an object called iris with the iris data
iris=load_iris()
# Print(iris)
# Creating a dataframe with four features variable
df=pd.DataFrame(iris.data,columns=iris.feature_names)
# Viewing the top 5 rows 
df.head()
```

```python
# Adding a new column for the species name
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
# viewing the top 5 rows
df.head()
# here we have added the target from iris dataset 
df.tail()
# here we can see dfferent flowers as output so therefore lets see how to train 
```

```python
# Creating Test and Train Data
df['is_train']=np.random.uniform(0,1,len(df)) <= 0.75 # the Data is divided into 75% training and 35% test
# Viewing the top 5 rows
df.head()
```

we can see that is train is showing the boolean value True.

```python
# Creating dataframe with test rows and train rows
train,test = df[df['is_train']==True],df[df['is_train']==False]
# show the number of observations for the test and train dataframes
print('Number of observations in the training data :',len(train))
print('Number of observations in the test data :',len(test))
```

```python
# Create a list of features column's names
features=df.columns[:4]
# View features
print(features)
```

```python
# Making the species into digital or numeric value so that ML understands
y= pd.factorize(train['species'])[0]
# viewing target
y
```

```python
# Prediction using Random Forest classifier
clf=RandomForestClassifier(n_jobs=2,random_state=0)
# Training the classifier
clf.fit(train[features],y) # fitting or training the data and then prediction sets so it tain the y , y is the target.
```

```python
# Applying the trained Classifier to the test 
clf.predict(test[features])
```

```python
# lets predict the data in another type 
# viewing the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
# here there are 3 flowers in matrix 
```

```python
# mapping names for the plants for each predictied plant class
preds=iris.target_names[clf.predict(test[features])]
# Viewing the Predicted species for the first five observations 
preds[0:5]
```

```python
# Viewing the Actual species for the first five observations 
test['species'].head()
```

```python
# Creating confusion matrix
# Cross tab take 2 sets of data and creates a chart out of them 
pd.crosstab(test['species'],preds,rownames=['Actual Species'],colnames=['predicted Species'])
```

Here we have 30 accurate values and 2 error so therefore we have almsot 93 % accuracy.
